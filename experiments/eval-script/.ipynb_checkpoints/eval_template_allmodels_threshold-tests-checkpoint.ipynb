```python
import matplotlib.pyplot as plt

def get_f1(clean_data, pseudo_data, generator_data, uniform_data):
    f1_data = []
    for i in range(len(clean_data)):
        TP = clean_data[i]
        TN = pseudo_data[i] + generator_data[i] + uniform_data[i]
        FP = 3 - TP
        FN = 1 - TP
        rec = TP/(TP+FN)
        prec = TP/(TP+FP)
        f1_data.append(2*rec*prec/(rec + prec))
    return f1_data
    

def model_plot(stats, model_name, figsize):
    #get data
    clean_data = []
    pseudo_data = []
    generator_data = []
    uniform_data = []
    
    thresholds_data = []
    
    for threshold in sorted(stats[model_name].keys()):
        thresholds_data.append(threshold)
        for data_type in stats[model_name][threshold].keys():
            if data_type == 'on_clean':
                clean_data.append(stats[model_name][threshold][data_type])
            if data_type == 'on_pseudo':
                pseudo_data.append(stats[model_name][threshold][data_type])
            if data_type == 'on_generator':
                generator_data.append(stats[model_name][threshold][data_type])
            elif data_type == 'on_uniform':
                uniform_data.append(stats[model_name][threshold][data_type])
    
    #get f1
    f1_data = get_f1(clean_data, pseudo_data, generator_data, uniform_data)
    print('f1:\t', f1_data)
    #plot
    plt.figure(figsize = figsize)
    plt.plot(thresholds_data, clean_data, label='100% clean accuracy')
    plt.plot(thresholds_data, pseudo_data, label='100% pseudo noise accuracy')
    plt.plot(thresholds_data, generator_data, label='100% generator noise accuracy')
    plt.plot(thresholds_data, uniform_data, label='100% uniform accuracy')
    plt.plot(thresholds_data, f1_data, label='f1 over all data')
    plt.title(model_name)
    plt.ylim(ymin=-0.2, ymax=1.2)
    plt.xlabel('Threshold value')
    plt.ylabel('Accuracy and F1')
    plt.legend()
    plt.show
```

```python
for model_path in model_paths:
    model_name = model_path[0]
    model_plot(stats, model_name, (8,4))
```
